#LINE_PLOT(MATPLOTLIB)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label="Sine Wave")
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

The above line plot visualizes the behavior of a sine wave over the range of 0 to 10 on the x-axis, with 100 evenly spaced values used as inputs for the sine function. The y-axis represents the corresponding sine values, which oscillate smoothly between -1 and 1, illustrating the periodic nature of the sine function. The plot includes a title labeled "Line Plot" and axes labeled "X-axis" and "Y-axis" to indicate the independent and dependent variables, respectively. A legend is also included, identifying the line as the "Sine Wave." This plot effectively demonstrates the cyclical pattern of the sine function, which repeats at regular intervals.

#SCATTER_PLOT(SEABORN)

```python
import seaborn as sns
import pandas as pd
import numpy as np

# Generate dataset
data = pd.DataFrame({
    'X': np.random.rand(50),
    'Y': np.random.rand(50)
})

sns.scatterplot(x='X', y='Y', data=data)
plt.title("Scatter Plot")
plt.show()
```

The above scatter plot visualizes the relationship between two randomly generated variables, X and Y, both containing 50 data points. Each point on the plot represents a unique pair of X and Y values, showing how they are distributed relative to one another. The x-axis corresponds to the values of the variable X, while the y-axis corresponds to the values of the variable Y. The title "Scatter Plot" is included to describe the visualization. In this case, the random generation of data suggests no clear relationship between X and Y.

#BAR_CHART(MATPLOTLIB)

```python
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

The bar chart above displays the values associated with four categories: A, B, C, and D. Each bar represents one category, with its height corresponding to the value associated with that category. The x-axis labels the categories, while the y-axis indicates the numerical values. Category A has a value of 23, category B has 45, category C has 56, and category D has the highest value at 78.

#HISTOGRAM(SEABORN)

```python
import seaborn as sns
import numpy as np

data = np.random.randn(100)

sns.histplot(data, kde=True, color='purple', bins=20)
plt.title("Histogram with KDE")
plt.show()
```

The histogram visualizes the distribution of a randomly generated dataset containing 100 data points. The x-axis represents the values of the dataset, while the y-axis indicates their frequency. The dataset is divided into 20 bins, as specified, with each bin showing the number of data points that fall within its range. 

Additionally, the chart includes a Kernel Density Estimate (KDE) curve, overlaid in purple, which smooths the data distribution and provides a continuous probability density function for the dataset. This KDE curve helps identify trends and patterns in the data, such as central tendencies and variability. The purple color enhances the chart's aesthetics, and the plot effectively combines the histogram and KDE to offer insights into the dataset's distribution.

#BOX_PLOT(SEABORN)

```python
import seaborn as sns
import pandas as pd

# Generate dataset
data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'] * 5,
    'Value': np.random.randn(20)
})

sns.boxplot(x='Category', y='Value', data=data)
plt.title("Boxplot")
plt.show()
```

The boxplot above visualizes the distribution of values for each of the four categories: A, B, C, and D. The x-axis represents the categories, while the y-axis shows the range of values associated with each category. 

For each category, the boxplot displays:
**The box**: It represents the interquartile range (IQR), which is the range between the 25th percentile (lower quartile) and the 75th percentile (upper quartile).
**The line inside the box**: This indicates the median, showing the central tendency of the values within the category.
**The whiskers**: These extend from the box to represent the range of data within 1.5 times the IQR from the quartiles, capturing the majority of the data points.
**Any outliers**: If present, they are shown as individual points outside the whiskers.

#HEATMAP(SEABORN)

```python
import seaborn as sns
import numpy as np

# Generate random data
data = np.random.rand(5, 5)

sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title("Heatmap")
plt.show()
```

The above heatmap provides a visual representation of a 5x5 matrix of randomly generated data, where each cell's color intensity corresponds to the magnitude of the data value in that position. 

Key aspects of the heatmap:
_Color Gradient_: The color map ("coolwarm") ranges from cool colors, blue for lower values, to warm colors red for higher values, making it easy to identify patterns or clusters in the data.
_Annotations_: The actual numerical values in each cell are displayed, offering precise information about the data at every position.
_Axes_: The x-axis and y-axis represent the rows and columns of the data matrix.

#PIE_CHART(MATPLOTLIB)

```python
labels = ['China', 'India', 'Nigeria', 'UK']
sizes = [10, 25, 60, 5]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Population of people living in poverty")
plt.show()
```

The above pie chart visually represents the distribution of populations living in poverty across four countries: China, India, Nigeria, and the UK. 

Key aspects of the chart:
_Proportions_: Each slice of the pie corresponds to the percentage of the total population living in poverty for each country. 
  **Nigeria** has the largest share (60%), indicating a significant portion of the population affected by poverty.
  **India** follows with 25%, representing a substantial but smaller proportion compared to Nigeria.
  **China** contributes 10%, showing a relatively smaller share.
  **The UK** accounts for only 5%, representing the smallest proportion.
_Labels and Percentages_: The chart labels each country and provides the exact percentage for easy interpretation. For example, 60.0% for Nigeria.
_Start Angle_: The chart starts at a 140-degree angle for better visual balance.

The chart highlights that Nigeria has the highest percentage of people living in poverty among the four countries, emphasizing the need for targeted interventions.

#3D_SCATTER_PLOT(MATPLOTLIB)


```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

ax.scatter(x, y, z, c='r', marker='o')
ax.set_title("3D Scatter Plot")
plt.show()
```

The above 3D scatter plot visualizes a set of data points distributed across three dimensions (x, y, and z). Each point represents a unique combination of values along these three axes. Here's the interpretation:

_Axes_:  
   - The **x-axis**, **y-axis**, and **z-axis** represent the three dimensions of the data, with values randomly generated between 0 and 1.
   - Each axis helps define the position of points in a 3D space.

_Data Points_:  
   - The plot contains 50 data points, each shown as a red marker (`c='r'` for red, `marker='o'` for circle).
   - The distribution of points demonstrates random scatter, emphasizing no specific trend or pattern.

_3D Perspective_:  
   - The use of 3D visualization helps provide depth and context, making it possible to observe the spatial relationship between data points in all three dimensions.
