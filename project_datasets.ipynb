#LINE_PLOT(MATPLOTLIB)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label="Sine Wave")
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

#SCATTER_PLOT(SEABORN)

```python
import seaborn as sns
import pandas as pd
import numpy as np

# Generate dataset
data = pd.DataFrame({
    'X': np.random.rand(50),
    'Y': np.random.rand(50)
})

sns.scatterplot(x='X', y='Y', data=data)
plt.title("Scatter Plot")
plt.show()
```

#BAR_CHART(MATPLOTLIB)

```python
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

#HISTOGRAM(SEABORN)

```python
import seaborn as sns
import numpy as np

data = np.random.randn(1000)

sns.histplot(data, kde=True, color='purple', bins=20)
plt.title("Histogram with KDE")
plt.show()
```

#BOX_PLOT(SEABORN)

```python
import seaborn as sns
import pandas as pd

# Generate dataset
data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'] * 5,
    'Value': np.random.randn(20)
})

sns.boxplot(x='Category', y='Value', data=data)
plt.title("Boxplot")
plt.show()
```

#HEATMAP(SEABORN)

```python
import seaborn as sns
import numpy as np

# Generate random data
data = np.random.rand(5, 5)

sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title("Heatmap")
plt.show()
```

#PIE_CHART(MATPLOTLIB)

```python
labels = ['A', 'B', 'C', 'D']
sizes = [25, 35, 20, 20]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart")
plt.show()
```

#3D_SCATTER_PLOT(MATPLOTLIB)

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

ax.scatter(x, y, z, c='r', marker='o')
ax.set_title("3D Scatter Plot")
plt.show()
```
